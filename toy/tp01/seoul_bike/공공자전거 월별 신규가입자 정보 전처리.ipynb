```python
df19 = pd.read_csv(path + '서울특별시 공공자전거 신규가입자 정보(월별)_2107_2112.csv', encoding='utf-8')
```

```python
def daytrans21(df):
    list21 = [202107,202108,202109,202110,202111,202112]
    for i in  list21:
        if i == 202107:
            df = df.replace(i, '21-Jul')
        elif i == 202108:
            df = df.replace(i, '21-Aug')
        elif i == 202109:
            df = df.replace(i, '21-Sep')
        elif i == 202110:
            df = df.replace(i, '21-Oct')
        elif i == 202111:
            df = df.replace(i, '21-Nov')
        elif i == 202112:
            df = df.replace(i, '21-Dec')
    return df
```

```python
df19 = daytrans21(df19)
```

```python
df19_2 = df19.groupby(['가입일시','연령대코드'])['가입 수'].sum()
```

```python
newdf = pd.DataFrame(columns= ['날짜', '10대','20대','30대','40대','50대','60대','70대','기타'])
```

```python
daylist21 = ['21-Jul','21-Aug','21-Sep','21-Oct','21-Nov','21-Dec']
daylist22 = ['22-Jan','22-Feb','22-Mar','22-Apr','22-May','22-Jun']
```

```python
for day in daylist21:
    dict_data = {'날짜':day, 
                 '10대':df19_2.loc[(day, '10대')],
                 '20대':df19_2.loc[(day, '20대')],
                 '30대':df19_2.loc[(day, '30대')],
                 '40대':df19_2.loc[(day, '40대')],
                 '50대':df19_2.loc[(day, '50대')],
                 '60대':df19_2.loc[(day, '60대')],
                 '70대':df19_2.loc[(day, '70대이상')],
                 '기타':df19_2.loc[(day, '기타')]}
    newdf = newdf.append(dict_data, ignore_index=True)
```

```python
def daytrans22(df):
    list22 = [202201,202202,202203,202204,202205,202206]
    for i in  list22:
        if i == 202201:
            df = df.replace(i, '22-Jan')
        elif i == 202202:
            df = df.replace(i, '22-Feb')
        elif i == 202203:
            df = df.replace(i, '22-Mar')
        elif i == 202204:
            df = df.replace(i, '22-Apr')
        elif i == 202205:
            df = df.replace(i, '22-May')
        elif i == 202206:
            df = df.replace(i, '22-Jun')
    return df
```
